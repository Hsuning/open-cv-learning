## Contrast and Brightness

### Calculation 
- g(x)=α*f(x)+β
- a(>0)
- β gain and offset value 
- contract : a+β
- brightness : β


- a = 1.68
- b = 125*(1-1.68)

```python
img = cv.imread("lena.jpg")

# a = 1.5, β=10
a = 1.68
b = 100*(1-1.68)
res = np.uint8(np.clip(a*img+b, 0, 255))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(img[:, :, ::-1])
axes[1].imshow(res[:, :, ::-1])
```
